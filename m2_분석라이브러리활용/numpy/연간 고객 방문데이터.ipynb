```python
import pandas as pd
import numpy as np
import random
from datetime import datetime
np.random.seed(0)
ar = np.random.rand(10000,13)

ard = pd.DataFrame(ar, index = range(1,10001), 
                   columns = ["cusno", "gender",'age','region','product_20','product_21','price_20','price_21','quantity_20','quantity_21','amount_20','amount_21','sales'])
ard.index.name = "TID"
ard.head()
```

```python
ard.tsid = np.arange(1,10001)
ard.cusno = np.random.randint(1000,2001,10000)
ard.gender = np.random.randint(0,2,10000)
ard.age = np.random.randint(0,81,10000)
ard.region = np.random.randint(1,11,10000)

start_date = pd.to_datetime("2020-01-01")
end_date = pd.to_datetime("2021-12-30")
ard["Datetime"] = pd.date_range(start ="2020-01-01" , end = "2021-12-30", periods = 10000) 
ard = ard.set_index(["Datetime", "tsid"])
```

```python
ard.price_20 = np.random.uniform(1000,5001,10000)
ard.price_21 = np.random.uniform(1000,5001,10000)
ard.quantity_20 = np.random.randint(1,101,10000)
ard.quantity_21 = np.random.randint(1,101,10000)

ard["amount_20"] = ard.price_20*ard.quantity_20
ard["amount_21"] = ard.price_21*ard.quantity_21
```

```python
ard["sales"] = ard['amount_21']>ard['amount_20']
ard['sales'].replace({True : 1, False : 0}, inplace = True)
```

```python
count, bin_dividers = np.histogram(ard.price_20, bins = 3)
bin_names = ['30','20','10']
ard.product_20 = pd.cut(x = ard.price_20,
                          bins=bin_dividers,
                          labels = bin_names,
                          include_lowest = True)
```

```python
count, bin_dividers = np.histogram(ard.price_21, bins = 3)
bin_names = ['30','20','10']
ard.product_21 = pd.cut(x = ard.price_21,
                          bins=bin_dividers,
                          labels = bin_names,
                          include_lowest = True)
```

```python
#1
count, bin_dividers = np.histogram(insite_df.amount_20, bins = 3)
bin_names = ['Bronz','Silver','Gold']
insite_df.product_20 = pd.cut(x = insite_df.amount_20,
                          bins=bin_dividers,
                          labels = bin_names,
                          include_lowest = True)
count, bin_dividers = np.histogram(insite_df.amount_21, bins = 3)
bin_names = ['Bronz','Silver','Gold']
insite_df.product_21 = pd.cut(x = insite_df.amount_21,
                          bins=bin_dividers,
                          labels = bin_names,
                          include_lowest = True)
```

```python
# 5
# 위 자료에서 볼 수 있듯이 전년도에 비해 브론즈 등급의 변화가 없는 고객이 전체의 37.7%정도이다.
# 브론즈 등급이 변화가 없거나 등급이 떨어진 고객을 'Question mark = ?'으로 정하겠다.
# 등급이 오른 고객은 'Star', 등급이 같은 고객은 'Cash_Cow'
```

```python
def bcg(x):
    if (x["Grade_20"] == "Bronz")&(x["Grade_21"] == "Bronz"):
        return "?"
    elif (x["Grade_20"] == "Silver")&(x["Grade_21"] == "Bronz"):
        return "?"
    elif (x["Grade_20"] == "Gold")&(x["Grade_21"] == "Bronz"):
        return "?"
    elif (x["Grade_20"] == "Gold")&(x["Grade_21"] == "Siver"):
        return "?"
    elif (x["Grade_20"] == "Gold")&(x["Grade_21"] == "Gold"):
        return "Cash_Cow"
    elif (x["Grade_20"] == "Silver")&(x["Grade_21"] == "Silver"):
        return "Cash_Cow"
    else:
        return "Star"
    
    
insite_df['BCG'] = insite_df.apply(lambda x: bcg(x), axis=1)
insite_df.head()
```

```python
insite_df['BCG'].value_counts()
```

```python
# 위 BCG 컬럼을 적용하여 
# 제품을 구매하지만 등급이 떨어지거나 제품 구매가 더딘 고객에겐 "?"마크를 줘서 이벤트를 열어 구매를 늘리면 혜택을 주는 마케팅을 사용하거나 
# 반대급부인 아무런 관심을 주지 않는 마케팅을 쓸 수도 있다. 
# 하지만 현 상황에서 대부분의 고객이 브론즈인 것을 감안해 고객들의 구매상황을 만들고 총구매금액을 올리는 적극적인 하위 고객 대상 마케팅이 필요하다.
# 이와 반대되는 고객들에겐 그에 맞는 맞춤형 마케팅을 선보일 수 있을 것이다.
```
