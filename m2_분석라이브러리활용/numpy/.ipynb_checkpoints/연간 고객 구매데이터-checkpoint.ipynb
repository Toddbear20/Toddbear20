```python
### import pandas as pd
import pandas as pd
import numpy as np
import random
from datetime import datetime
np.random.seed(4)
ar = np.random.rand(10000,8)

ard = pd.DataFrame(ar, index = range(1,10001), 
                   columns = ["cusno", "gender",'age','region','product','price','quantity','amount'])
ard.index.name = "TID"
ard.head()
```

```python

from datetime import datetime
ard.cusno = np.random.randint(1000,2001,10000)
ard.gender = np.random.randint(0,2,10000)
ard.age = np.random.randint(0,81,10000)
ard.region = np.random.randint(1,11,10000)
date_list= list(pd.date_range(start ="2020-01-01" , end = "2021-12-31", freq = 'D'))
ard['Date'] = np.random.choice(date_list,size = 10000)
ard['day']=ard['Date'].apply(lambda x: x.weekday())  #date별 요일을 숫자로 나타냄a
ard['day']=ard['day'].replace(list(range(7)),['월','화','수','목','금','토','일'])
ard['time']=np.random.randint(1,25,size=10000)
```

```python
ard.price = np.random.uniform(1000,5001,10000)
ard.quantity = np.random.randint(1,101,10000)
ard["amount"] = ard.price*ard.quantity
```

```python
count, bin_dividers = np.histogram(ard.price, bins = 3)
bin_names = ['30','20','10']
ard["product"] = pd.cut(x = ard.price,
                          bins=bin_dividers,
                          labels = bin_names,
                          include_lowest = True)
```
