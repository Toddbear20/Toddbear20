```python
# 자치구 단위 내 서울 생활인구 데이터 불러오기
import pandas as pd
df = pd.read_csv('./자치구 단위 서울 생활인구(내국인).csv', encoding='cp949')
df
```

```python
# 자치구코드로 마포구를 찾기/마포구 자치구코드 = 11440
df_mapo = df[df['자치구코드']==11440]
df_mapo
```

```python
from datetime import datetime
import warnings
warnings.filterwarnings(action='ignore')
df_mapo['기준일ID'] = df_mapo['기준일ID'].apply(str)
```

```python
df_mapo.info()
```

```python

df_mapo['기준일ID']= pd.to_datetime(df_mapo['기준일ID'],format='%Y%m%d')
df_mapo['Year']=df_mapo['기준일ID'].dt.year
df_mapo['Month']=df_mapo['기준일ID'].dt.month
df_mapo['Day']=df_mapo['기준일ID'].dt.day
```

```python
df_2022 = df_mapo[(df_mapo['Year']==2022)&(df_mapo['Month']==7)&(df_mapo['Day']== 24)]
```
