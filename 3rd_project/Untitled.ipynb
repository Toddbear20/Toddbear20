```python
training_df['출하량'] = np.log1p(training_df['출하량'])

training_df['상대습도'] = np.log1p(training_df['상대습도'])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# 모델 생성
# lr_clf = LogisticRegression()
lr_clf = LinearRegression()

# 피쳐, 타겟 지정
x_feature = training_df.drop(['수주량'],axis=1)
y_label = training_df['수주량'].values

x_train,x_test,y_train,y_test = train_test_split(x_feature,y_label,test_size=0.2, random_state=10)

lr_clf.fit(x_train,y_train)
pred=lr_clf.predict(x_test)

mse = mean_squared_error(y_test,pred)
mae = mean_absolute_error(y_test,pred)
r2 = r2_score(y_test,pred)

print(np.sqrt(mse))
print(mae)
print(r2)
```
