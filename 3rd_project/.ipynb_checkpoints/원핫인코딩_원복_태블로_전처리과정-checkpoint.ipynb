```python
import pandas as pd

training_df = pd.read_csv('./학습용 데이터셋_2년.csv', index_col=0)
training_df
```

```python
training_df.columns
```

```python
pro_li = ['AE', 'ARE-580FX', 'CSA4000', 'CSA5000', 'PEMA-500FR', 'PEMA-580FX',
       'PEMA-CR1000', 'PEMA-CSA5000', 'PEMA-HR1000', 'PEMA-HR1000S',
       'PEMA-HR1500', 'PEMA-PCM2000', 'PEMA-PCM2000B', 'PEMA-PCR3000E',
       'PEMA-PCR3000N', 'PEMA-PR1000', 'PEMA-PR2000', 'PEMA-SN400',
       'PEMA-SP1000', 'PEMA-SPR', 'PEMA-SR2000', 'PEMA-SR2000A',
       'PEMA-SR3000F', 'PEMA-SR5000F', 'PR1000', 'SRE-110', 'SRE-200']
```

```python
training_df['제품명'] = 0
for i in pro_li:
    training_df['제품명'].loc[training_df[i]==1] = i
```

```python
training_df.to_csv('학습용데이터_2년.csv')
```
