```python
import re
import pandas as pd
def yearmonth(x):
    p = re.compile('.{7}')
    return p.findall(x)[0]
```

```python

weather_df['Month'] = weather_df['날짜'].apply(yearmonth)
weather_df
```

```python
weather_pv = weather_df.pivot_table(index = 'Month',
                           values='평균온도',
                           aggfunc='mean')
weather_pv.reset_index(inplace=True)
```

```python
weather_pv_1 = weather_df.pivot_table(index = 'Month',
                           values='일강수량',
                           aggfunc='mean')
weather_pv_1.reset_index(inplace=True)
```

```python
weather_pv_2 = weather_df.pivot_table(index = 'Month',
                           values='상대습도',
                           aggfunc='mean')
weather_pv_2.reset_index(inplace=True)
```

```python
weather_pv_3 = weather_df.pivot_table(index = 'Month',
                           values='최심적설',
                           aggfunc='mean')
weather_pv_3.reset_index(inplace=True)
```

```python
from functools import reduce

df_li = [money_flow2,build_df,weather_pv,weather_pv_1,weather_pv_2,weather_pv_3]

feature_df = reduce(lambda x,y : pd.merge(x,y, on='Month', how='outer'), df_li)
```
