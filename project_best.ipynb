```python
# 맛집 검색 및 별점, 방문자리뷰 수 크롤링
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import requests
from bs4 import BeautifulSoup
import re
import csv
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

path = 'C:/Users/sch014/working/git_mine/Toddbear20/m2_분석라이브러리활용/crawling/chromedriver.exe'
options = webdriver.ChromeOptions()

driver = webdriver.Chrome(path, options=options)
driver.implicitly_wait(time_to_wait=3)

driver.get("https://map.naver.com/v5/") 
try:
    element = WebDriverWait(driver, 10).until(
       EC.presence_of_element_located((By.CLASS_NAME, "input_search"))
   ) #입력창이 뜰 때까지 대기
finally:
    pass

search_box = driver.find_element(By.CLASS_NAME,"input_search")
search_box.send_keys("이대역맛집")
search_box.send_keys(Keys.ENTER)
time.sleep(4)

frame = driver.find_element(By.CSS_SELECTOR,"iframe#searchIframe")

driver.switch_to.frame(frame)

time.sleep(3)
# 여기까지 iframe 전환

scroll_div = driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/div[1]")
#검색 결과로 나타나는 scroll-bar 포함한 div 잡고
driver.execute_script("arguments[0].scrollBy(0,2000)", scroll_div)
time.sleep(2)
driver.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
time.sleep(2)
driver.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
time.sleep(2)
driver.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
time.sleep(2)
driver.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
time.sleep(2)

final_result = []

i = 2
while i<=5: #몇 페이지까지 크롤링할 것인지 지정
    stores = driver.find_elements(By.CSS_SELECTOR,"#_pcmap_list_scroll_container > ul > li")
   #해당 페이지에서 표시된 모든 가게 정보
   
    for store in stores: #한 페이지 내에서의 반복문. 순차적으로 가게 정보에 접근
        name = store.find_element(By.CSS_SELECTOR,"#_pcmap_list_scroll_container > ul > li> div > a > div > div > span.place_bluelink.OXiLu").text #가게 이름
        
        try:    
            rating = re.search('(\d.\d+)', store.find_element(By.CSS_SELECTOR,"#_pcmap_list_scroll_container > ul > li > div > div > span > em").get_attribute('innerHTML')).groups()[0]
        except:
            rating = ''
        try:
            visit = re.sub('[^\d+,]','',store.find_element(By.CSS_SELECTOR,'#_pcmap_list_scroll_container > ul > li > div._3hn9q > div._17H46 > span:nth-child(3)').text)
        except:
            visit = ''
        
#         click_name = store.find_element(By.CSS_SELECTOR,"#_pcmap_list_scroll_container > ul > li > div > a > div > div > span.place_bluelink.OXiLu")
#         click_name.click() 
        
#         driver.switch_to.default_content()
#         time.sleep(4)  
        
#         frame_in = driver.find_element(By.XPATH,'/html/body/app/layout/div[3]/div[2]/shrinkable-layout/div/app-base/search-layout/div[2]/entry-layout/entry-place-bridge/div/nm-external-frame-bridge/nm-iframe/iframe')
#         driver.switch_to.frame(frame_in) 
#         time.sleep(3)
        
#         try:
#             address = re.search('>[서울\w+]<', driver.find_element(By.CSS_SELECTOR,"#app-root > div > div > div > div > div > div > div > ul > li > div > a > span._2yqUQ").text)
#         except:
#             address = ''
#         try:
#             visit = re.search('(\d+)',store.find_element(By.CSS_SELECTOR,'#app-root > div > div > div > div > div > div > span > a > em').get_attribute('innerHTML')).groups()[0]
#         except:
#             visit = ''
        
        store_info = {
           'place':name,
           'rate':rating, 'visit':visit}
        
        final_result.append(store_info)
        
#         driver.switch_to.default_content()
#         driver.switch_to.frame(frame)
        
        time.sleep(3)
     
    next_button = driver.find_element(By.LINK_TEXT, str(i))
    next_button.click()
    i = i+1
    time.sleep(5)
    
print(final_result)

    
```
